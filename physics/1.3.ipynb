```python
import numpy as np
import matplotlib.pyplot as plt
from math import sqrt, pi
%matplotlib inline
```

Исследование зависимости B = B(H).

```python
N1 = 100
N2 = 200
r = 1.35
R1 = 51
R2 = 4300
C = 0.22
S = 0.19625
```

```python
Ux = np.array([0.08, 0.06, 0.048, 0.04, 0.034, 0.026, 0.0215, 0.017, 0.0115])
Uy = np.array([1.5, 1.350, 1.25, 1.1, 0.95, 0.8, 0.64, 0.48, 0.32])
```

```python
H = (2 * N1 * Ux) / (10 * r * R1)
B = (100 * R2 * C * Uy) / (N2 * S)
```

```python
fig = plt.figure(figsize=(15,7))
ax = fig.gca()
ax.set_xticks(np.arange(0, 0.024, 0.00025), minor=True)
ax.set_yticks(np.arange(0, 3700, 50), minor=True)
ax.grid(which='minor', alpha=0.4, linestyle='-')

plt.plot(H, B, c='blue', linewidth=2)
plt.scatter(H, B, c='black', linewidth=3)
plt.xlabel("H, СГСЭ", horizontalalignment='right', fontsize='x-large', x=1)
h = plt.ylabel(r"B, СГСЭ", verticalalignment='top', fontsize='x-large', horizontalalignment='right', y=1)
h.set_rotation(0)

plt.title('Зависимость B от H', fontsize=16)
plt.grid(ls='--')
plt.savefig("First")
plt.show()
```

```python
fig = plt.figure(figsize=(15,7))
ax = fig.gca()
ax.set_xticks(np.arange(0, 0.024, 0.00025), minor=True)
ax.set_yticks(np.arange(0, 260000, 2000), minor=True)
ax.grid(which='minor', alpha=0.4, linestyle='-')

plt.plot(H, B / H, c='blue', linewidth=2)
plt.scatter(H, B / H, c='black', linewidth=3)
plt.xlabel("H, СГСЭ", horizontalalignment='right', fontsize='x-large', x=1)
h = plt.ylabel(r"$\mu$", verticalalignment='top', fontsize='x-large', horizontalalignment='right', y=1)
h.set_rotation(0)

plt.title('Зависимость $\mu$ от H', fontsize=16)
plt.grid(ls='--')
plt.savefig("Second")
plt.show()
```

```python
fig = plt.figure(figsize=(15,7))
ax = fig.gca()
ax.set_xticks(np.arange(0, 0.024, 0.00025), minor=True)
ax.set_yticks(np.arange(0, 300, 5), minor=True)
ax.grid(which='minor', alpha=0.4, linestyle='-')

plt.plot(H, (B - H) / (4 * pi), c='blue', linewidth=2)
plt.scatter(H, (B - H) / (4 * pi), c='black', linewidth=3)
plt.xlabel("H, СГСЭ", horizontalalignment='right', fontsize='x-large', x=1)
h = plt.ylabel(r"M, СГСЭ", verticalalignment='top', fontsize='x-large', horizontalalignment='right', y=1)
h.set_rotation(0)

plt.title('Зависимость M от H', fontsize=16)
plt.grid(ls='--')
plt.savefig("Third")

plt.show()
```

Исследование зависимости D = D(E).

```python
h = 0.1
C = 0.047
S = 0.21
```

```python
Ux = np.array([2.6, 2.2, 2, 1.7, 1.4, 1.15, 0.86, 0.56, 0.29, 0.11])
Uy = np.array([12.5, 10.5, 6.8, 2.8, 1.05, 0.46, 0.26, 0.15, 0.018, 0.0036])
U = np.array([72.01, 64., 56.13, 47.83, 40.17, 32.05, 24.15, 16., 8.09, 0.62])
```

```python
E = (sqrt(2) * U) / (300 * h)
D = (12 * pi * 1000 * C * Uy) / S
```

```python
fig = plt.figure(figsize=(15,7))
ax = fig.gca()
ax.set_xticks(np.arange(0, 3.5, 0.05), minor=True)
ax.set_yticks(np.arange(0, 110000, 2500), minor=True)
ax.grid(which='minor', alpha=0.4, linestyle='-')

plt.plot(E, D, c='blue', linewidth=2)
plt.scatter(E, D, c='black', linewidth=3)
plt.xlabel("E, СГСЭ", horizontalalignment='right', fontsize='x-large', x=1)
h = plt.ylabel(r"D, СГСЭ", verticalalignment='top', fontsize='x-large', horizontalalignment='right', y=1)
h.set_rotation(0)

plt.title('Зависимость D от E', fontsize=16)
plt.grid(ls='--')
plt.savefig("Forth")
plt.show()
```

```python
fig = plt.figure(figsize=(15,7))
ax = fig.gca()
ax.set_xticks(np.arange(0, 3.5, 0.05), minor=True)
ax.set_yticks(np.arange(0, 33000, 1000), minor=True)
ax.grid(which='minor', alpha=0.4, linestyle='-')

plt.plot(E, D / E, c='blue', linewidth=2)
plt.scatter(E, D / E, c='black', linewidth=3)
plt.xlabel("E, СГСЭ", horizontalalignment='right', fontsize='x-large', x=1)
h = plt.ylabel(r"$\epsilon$", verticalalignment='top', fontsize='x-large', horizontalalignment='right', y=1)
h.set_rotation(0)

plt.title(r'Зависимость $\epsilon$ от E', fontsize=16)
plt.grid(ls='--')
plt.savefig("Fifth")
plt.show()
```

```python
fig = plt.figure(figsize=(15,7))
ax = fig.gca()
ax.set_xticks(np.arange(0, 3.5, 0.05), minor=True)
ax.set_yticks(np.arange(0, 8700, 250), minor=True)
ax.grid(which='minor', alpha=0.4, linestyle='-')

plt.plot(E, (D - E) / (4 * pi), c='blue', linewidth=2)
plt.scatter(E, (D - E) / (4 * pi), c='black', linewidth=3)
plt.xlabel("E, СГСЭ", horizontalalignment='right', fontsize='x-large', x=1)
h = plt.ylabel(r"P, СГСЭ", verticalalignment='top', fontsize='x-large', horizontalalignment='right', y=1)
h.set_rotation(0)

plt.title(r'Зависимость P от E', fontsize=16)
plt.grid(ls='--')
plt.savefig('Sixth')
plt.show()
```
