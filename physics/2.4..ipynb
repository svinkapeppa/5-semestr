```python
import numpy as np
import matplotlib.pyplot as plt
from math import sqrt
%matplotlib inline
```

```python
def getk(x, y):
    return np.mean(x * y) / np.mean(x ** 2)

def geterr(x, y, n):
    return sqrt(((np.mean(y ** 2))/(np.mean(x ** 2)) - (getk(x, y) ** 2))) / sqrt(n)

# TODO: change getk and geterr that they can work both for y = kx and y = kx + b
```

** Измерения на первой трубе **

```python
P_KG = np.array([12, 24, 38, 62, 114, 149, 184, 226, 259])
P = P_KG * 0.2 * 9.80665
P = np.insert(P, 0, 0)
```

```python
V = np.array([1, 2, 3])
```

```python
T_1 = np.array([[61, 61], [126, 126], [189, 189]])
t_1 = list(map(lambda t: np.mean(t), T_1))
T_2 = np.array([[30, 30], [60, 61], [91, 92]])
t_2 = list(map(lambda t: np.mean(t), T_2))
T_3 = np.array([[19, 20], [39, 39], [58, 59]])
t_3 = list(map(lambda t: np.mean(t), T_3))
T_4 = np.array([[12.63, 12.63], [25.44, 25.69], [38.16, 37.21]])
t_4 = list(map(lambda t: np.mean(t), T_4))
T_5 = np.array([[9.60, 9.56], [18.94, 19.09], [29.19, 29.19]])
t_5 = list(map(lambda t: np.mean(t), T_5))
T_6 = np.array([[8.63, 8.66], [17.56, 17.28], [26.15, 26.0]])
t_6 = list(map(lambda t: np.mean(t), T_6))
T_7 = np.array([[8.16, 8.06], [16.0, 15.78], [23.75, 23.50]])
t_7 = list(map(lambda t: np.mean(t), T_7))
T_8 = np.array([[6.87, 6.62], [14.5, 14.37], [21.03, 21.28]])
t_8 = list(map(lambda t: np.mean(t), T_8))
T_9 = np.array([[6.53, 6.55], [13.22, 13.37], [19.78, 19.91]])
t_9 = list(map(lambda t: np.mean(t), T_9))

t = []
t.append(t_1)
t.append(t_2)
t.append(t_3)
t.append(t_4)
t.append(t_5)
t.append(t_6)
t.append(t_7)
t.append(t_8)
t.append(t_9)
```

```python
qs = list(map(lambda y: V / y, t))
q = list(map(lambda y: np.mean(y), qs))
q = np.insert(q, 0, 0)
```

```python
k_1 = getk(q[:5], P[:5])
```

```python
fig = plt.figure(figsize=(15,15))
ax = fig.gca()
ax.set_xticks(np.arange(0, 0.2, 0.002), minor=True)
ax.set_yticks(np.arange(0, 500, 5), minor=True)
ax.grid(which='minor', alpha=0.4, linestyle='-')

plt.plot(q[:5], k_1*np.array(q[:5]), c='black', linewidth=2)
plt.scatter(q, P, c='r', linewidth=5)

plt.xlabel(r'$\Delta Q\ [\frac{м}{с} \cdot 10^{-3}]$', horizontalalignment='right', fontsize='x-large', x=1)
h = plt.ylabel(r"$\Delta P\ [Па]$", verticalalignment='top', fontsize='x-large', horizontalalignment='right', y=1)
h.set_rotation(0)

plt.title('Зависимость $\Delta P$ от $\Delta Q$')
plt.grid()
plt.savefig("First")
plt.show()
```
