```python
import numpy as np
import matplotlib.pyplot as plt
from math import sqrt
%matplotlib inline
```

```python
def getk(x, y):
    return np.mean(x * y) / np.mean(x ** 2)

def geterr(x, y, n):
    return sqrt(((np.mean(y ** 2))/(np.mean(x ** 2)) - (getk(x, y) ** 2))) / sqrt(n)
```

Измерения на первой установке.

```python
x_1 = np.array([0, 1, 2, 3, 4])
y_1 = np.array([0, 4.8, 9.7, 14.4, 19.2])
k_1 = getk(x_1, y_1)
err_1 = geterr(x_1, y_1, 4)
```

```python
fig = plt.figure(figsize=(15,15))
ax = fig.gca()
ax.set_xticks(np.arange(0, 4, 0.025), minor=True)
ax.set_yticks(np.arange(0, 20, 0.1), minor=True)
ax.grid(which='minor', alpha=0.4, linestyle='-')
plt.plot(x_1, k_1*x_1, c='black', linewidth=2)
plt.scatter(x_1, y_1, c='r', linewidth=5)
plt.xlabel("№", horizontalalignment='right', fontsize='x-large', x=1)
h = plt.ylabel(r"м, $10^{-2}$", verticalalignment='top', fontsize='x-large', horizontalalignment='right', y=1)
h.set_rotation(0)
plt.grid()
plt.savefig("First")
plt.show()
```
