```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
worst = []
mid = []
best = []
data = open('data.txt', 'r')

for line in data:
    line = line.split(" ")
    if(line[0] == '\n'):
        continue
    best.append(float(line[0]))
    worst.append(float(line[1]))
    mid.append(float(line[2]))
data.close()

t = np.arange(1, len(worst) + 1)

plt.figure(figsize=(10, 5))

plt.plot(t, worst, label='worst')
plt.plot(t, mid, label='average')
plt.plot(t, best, label='best')

plt.title('N = 10000')
plt.legend()
plt.grid(ls='--')
plt.show()
```
