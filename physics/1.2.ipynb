```python
import numpy as np
import matplotlib.pyplot as plt
from math import sqrt
%matplotlib inline
```

```python
def getk(x, y):
    return np.mean(x * y) / np.mean(x ** 2)

def geterr(x, y, n):
    return sqrt(((np.mean(y ** 2))/(np.mean(x ** 2)) - (getk(x, y) ** 2))) / sqrt(n)
```

```python
z = np.linspace(0, 15, 31)
B = np.array([7.84, 7.94, 8.04, 8.1, 8.16, 8.19, 8.22, 8.24, 8.25,
              8.24, 8.21, 8.19, 8.14, 8.09, 8.02, 7.93, 7.81, 7.67,
              7.49, 7.18, 6.85, 6.44, 5.92, 5.32, 4.63, 3.91, 3.18,
              2.56, 2.14, 1.71, 1.3])
```

```python
fig = plt.figure(figsize=(15,7))
ax = fig.gca()
ax.set_xticks(np.arange(0, 16, 0.2), minor=True)
ax.set_yticks(np.arange(1, 9, 0.2), minor=True)
ax.grid(which='minor', alpha=0.4, linestyle='-')

plt.plot(z, B, c='blue', linewidth=2)
plt.scatter(z, B, c='black', linewidth=3)
plt.xlabel("z, cm", horizontalalignment='right', fontsize='x-large', x=1)
h = plt.ylabel(r"B, mT", verticalalignment='top', fontsize='x-large', horizontalalignment='right', y=1)
h.set_rotation(0)

plt.title('Зависимость B от z', fontsize=16)
plt.grid(ls='--')
plt.savefig("First")
plt.show()
```

```python
I = np.linspace(0.5, 2.5, 5)
B = np.array([8.25, 15.84, 23.2, 30.4, 38.1])
k = getk(I, B)
```

```python
fig = plt.figure(figsize=(15,7))
ax = fig.gca()
ax.set_xticks(np.arange(0, 3, 0.025), minor=True)
ax.set_yticks(np.arange(8, 40, 0.5), minor=True)
ax.grid(which='minor', alpha=0.4, linestyle='-')

plt.plot(I, k * I, c='blue', linewidth=2)
plt.scatter(I, B, c='black', linewidth=3)
plt.xlabel("I, A", horizontalalignment='right', fontsize='x-large', x=1)
h = plt.ylabel("B, mT", verticalalignment='top', fontsize='x-large', horizontalalignment='right', y=1)
h.set_rotation(0)

plt.title('Зависимость B от I', fontsize=16)
plt.grid(ls='--')
plt.savefig("Second")
plt.show()
```
