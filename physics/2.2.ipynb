```python
fig = plt.figure(figsize=(15,7))

ax = fig.gca()
ax.set_xticks(np.arange(2, 16, 0.2), minor=True)
ax.set_yticks(np.arange(0., 6., 0.2), minor=True)
ax.grid(which='minor', alpha=0.4, linestyle='-')

plt.scatter(f[0], i[0], c='r', marker='^', label='R = 1 Ом')
plt.plot(f[0], i[0], color='r', label='Апроксимация для R = 1 Ом')
plt.scatter(f[1], i[1], c='g', marker='o', label='R = 500 Ом')
plt.plot(f[1], i[1], color='g', label='Апроксимация для R = 500 Ом')
plt.scatter(f[2], i[2], c='b', marker='v', label='R = 3000 Ом')
plt.plot(f[2], i[2], color='b', label='Апроксимация для R = 3000 Ом')

plt.xlabel(r'$f, kHz$', horizontalalignment='right', fontsize='x-large', x=1)
h = plt.ylabel(r'$I, mA$', verticalalignment='top', fontsize='x-large', horizontalalignment='right', y=1)
h.set_rotation(0)

plt.title('Зависимость I от f', fontsize=16)
plt.legend(fontsize=16)
plt.grid(ls='--')
plt.show()
```
